#### Answer_3

##### > ArithmeticError class consist of all those error which occurs due to               ivalid arithmetic operation.
##### > ArithmeticError class consist of following errors:
###### > OverflowError
###### > ZeroDivisionError
###### > FloatingPointError

```python
## Example ZeroDivision Error

try:
    a = 6/0
except ZeroDivisionError as e:
    print("This error is part of ArithmeticError class which is:",e)
```

```python
## Example OverflowError
## It occurs when arithmetic operation exceeds a limit
j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
        print(j)
except OverflowError as e:
    print("Overflow error happened")
    print(f"{e}, {e.__class__}")
```

#### Answer_4

##### > The LookupError exception in Python forms the base class for all                   exceptions that are raised when an index or a key is not found for a               sequence or dictionary respectively.
##### > We can use LookupError exception class to handle both IndexError and               KeyError exception classes.

```python
## IndexError Example

lst = [1,2,3,4,5]

try:
    print(lst[5])
except LookupError as e:
    print(f"{e} ,{e.__class__}")
```

```python
## KeyError Example

dic = {"key_1" : "value_1","key_2" : "value_2"}

try:
    print(dic["key_3"])
except LookupError as e:
    print(f"{e} , {e.__class__}")
```
